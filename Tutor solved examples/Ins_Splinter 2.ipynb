```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Set up Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Visit the Quotes to Scrape site
url = 'http://quotes.toscrape.com/'
browser.visit(url)
```

```python
# Parse the HTML
html = browser.html
html_soup = soup(html, 'html.parser')
```

```python
# Scrape the Title
title = html_soup.find('h2').text
title
```

```python
# Scrape the top ten tags
tag_box = html_soup.find('div', class_='tags-box')
# tag_box
tags = tag_box.find_all('a', class_='tag')

for tag in tags:
    word = tag.text
    print(word)
```

```python
for x in range(1, 6):
    html = browser.html
    quote_soup = soup(html, 'html.parser')
    quotes = quote_soup.find_all('span', class_='text')
    
    for quote in quotes:
        print('page:', x, '----------')
        print(quote.text)
    
    browser.links.find_by_partial_text('Next').click()
```

```python
for x in range(1, 6):
    html = browser.html
    quote_soup = soup(html, 'html.parser')
    quotes = quote_soup.find_all('div', class_='quote')
    
    for quote in quotes:
        print('page:', x, '----------')
        print(quote.text)
    
    browser.links.find_by_partial_text('Next').click()
```

### Skill Drill: Books to Scrape

```python
import pandas as pd

# Visit the Books to Scrape site
url = 'http://books.toscrape.com/'
browser.visit(url)
```

```python
html = browser.html
html_soup = soup(html, 'html.parser')
```

```python
link_set = []
title_set = []

links = html_soup.find_all('a', href=True)

for link in links:
    try:
        title_set.append(link['title'])
        link_set.append(url+link['href'])
    except:
        continue

# Create a dataframe from the link_set and title_set, combined using zip() [more on zip below]        
link_df = pd.DataFrame(zip(title_set, link_set), columns=['Title', 'Link'])

link_df
```

#### A little more on zip(), if you want more

```python
# If you haven't seen much of the zip() function in action yet, here's an example.
# Zip() takes an arbitrary number of equal-sized iterables (ranges, lists, tuples, dictionaries, etc.), lines them
#   up element-by-element and combines them, e.g., first item in each iterable together as a tuple, then second item
#   in each iterable together as a tuple, etc.

def fizzbuzz(n):
    str_fizz = ''
    str_buzz = ''
    if n%3==0: 
        str_fizz = 'Fizz'
    if n%5==0:
        str_buzz = 'Buzz'
    return str_fizz + str_buzz
        

first_range = range(1,20)
second_range = [n*2 for n in first_range]
mod_3 = [fizzbuzz(n) for n in second_range]



for i in zip(first_range, second_range, mod_3):
    print(i)
```
